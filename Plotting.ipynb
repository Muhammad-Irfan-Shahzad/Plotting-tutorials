# All types of plotting use in Data Science

```python
#import the boston dataset from sklearn library
from sklearn.datasets import load_boston
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#load boston dataset
boston_real_state_data = load_boston()
```

```python
# view boston dataset
print(boston_real_state_data.DESCR)
```

# Histogram

```python
# define x axis for the data
x_axis = boston_real_state_data.data
```

```python
# define y axis for the target
y_axis = boston_real_state_data.target
```

```python
# Histogram plot
style.use('ggplot') 
plt.figure(figsize = (7,7)) 
plt.hist(y_axis,bins =50)
plt.xlabel('price in 1000s USD')
plt.ylabel('number of house')
plt.show()
```

# Scatter Plot

```python
# Scatter plot
style.use('ggplot') 
plt.figure(figsize = (7,7)) 
plt.scatter(boston_real_state_data.data[:,5],boston_real_state_data.target)
plt.xlabel('price in 1000s USD')
plt.ylabel('number of house')
plt.show()
```

# Heatmap

```python
import seaborn as sns
```

# Pie Chart

```python
# job data in percentile
job_data = ['40','20','17','8','5','10']
#define label as different departments
labels = 'IT', 'Finance','Marketing','Admin','HR','Operations'
# explode the 1st slice which is IT
explode = (0.05,0,0,0,0,0)
#draw the pie chart and set the paramters
plt.pie(job_data,labels = labels,explode = explode)
#show the plot
plt.show()
```

```python
#X-Y plot
web_traffic = [1,2,3,4,5,6]
time_hrs =[23,53,64,17,17,24]
style.use('ggplot')
plt.plot(web_traffic,time_hrs,linestyle='--',color='b',label='Web Traffic')
plt.xlabel('Web traffic')
plt.ylabel('Time hrsv')
plt.title('Website Traffic analysis')
plt.legend()
plt.show()
```

# Pair Plot

```python
import pandas as pd
#import the auto data set
df_auto_dataset = pd.read_csv('auto_data-mpg.csv')
df_auto_dataset.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
# view top 5 record
df_auto_dataset.head()
```

```python
#write a user defined function for orgin
# 1-USA, 2-Europe, 3-Asia
def orgin(num):
    if num==1:
        return 'USA'
    elif(num==2):
        return 'Europe'
    else:
        return 'Asia'
#use apply function
df_auto_dataset['orgin'] = df_auto_dataset['orgin'].apply(orgin) 
```

```python
# view top 5 record
df_auto_dataset.head()
```

```python
# view the first 30 record
df_auto_dataset.head(30)
```

```python
#draw the pair plot using sns for mpg, weight,orgin and with hue origin, set the size to 4
#note: hue is variable in data set to map plot aspects to different colors
sns.pairplot(df_auto_dataset[['mpg','weight','orgin']],hue='orgin', size=4)
```

# Error Bar

```python
# Example data
x = np.linspace(0, 10, 11)
y = x ** 2
y_err = 2 * x

# Plot the data
plt.errorbar(x, y, yerr=y_err, fmt='o')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error Bar Plot')

# Show the plot
plt.show()
```
